# universal funksiyalar.

Unary funksiyalar ( 1 ta argument qabul qiluvchi funksiyalar.)

sqrt , square , exp , log , modf , sign , isnan funksiyalari bilan tanishamiz.

```python
arr1 = np.arange(10)
arr1
```

```python
np.sqrt(arr1) # bu funksiya massivning har bir elementini kv ildizini qaytaradi.
```

```python
np.square(arr1) # square() funksiyasi massiv elementlarining kvadratini qaytaradi.
```

```python
np.exp(arr1) # bu funksiya har bir elementning exponentini qaytaradi.
```

```python
np.log(np.exp(arr1)) # bu funksiya e darajali logorifmda har bir element qiymatini qaytaradi.
```

```python
np.log2(4)
```

```python
arr2 = np.random.randn(10)
arr2
```

```python
qoldiq_m , butun_m = np.modf(arr2) # bu funksiya har bir elementning butun va qoldiq qismlarini alohida massivlarga joylaydi.
```

```python
print(qoldiq_m)
print(butun_m)
```

```python
np.modf(arr2)
```

```python
np.sign(arr2)
```

```python
arr3 = np.array([-12,34,1,5,-7,54,-234])
```

```python
np.sign(arr3) # bu funksiya massiv elementlarining ishoralarini qaytaradi.
```

```python
arr3 = np.float64(arr3)
arr3[4:] = np.nan
```

```python
np.isnan(arr3) # bu funksiya massivdagi nan qiymatlarni aniqlaydi va ular o'rniga true qolganlariga esa false qiymat yuklab massiv qaytaradi.
```

abs, fabs, ceil, floor va rint funksiyalari.

```python
np.abs([3+4j])     # [5.] bu funksiya massiv ichidagi har bir elementning modulini qaytaradi. istalgan turdagi sonlar uchun: int, float, complex
```

```python
np.fabs([-12,4]) # faqat real sonlar bilan ishlaydi farqi shunda va faqat float qiymat qaytaradi.
```

```python
np.ceil([-1.2, -2.8, 1.1, 3.7]) # bu funksiya sonning o'zini yoki o'zidan katta bo'lgan butun sonni qaytaradi.
```

```python
np.floor([-1.2, -2.8, 1.1, 3.7]) # bu funksiya sonning o'zini yoki o'zidan kichik bo'lgan butun sonni qaytaradi.
```

```python
np.rint([-1.2, -2.8, 1.1, 3.7]) # sonning o'zini yoki o'ziga eng yaqin bo'lgan butun sonni qaytaradi.
```

## Binary funksiyalar-ikkita argument qabul qiluvchi funksiyalar.

add, multiply, maximum  funksiyalarini ko'rib chiqamiz.

```python
arr6 = np.random.randn(7)
arr5 = np.random.randn(7)
```

```python
arr6
```

```python
arr5
```

```python
np.add(arr5,arr6) # add funksiyasi massivlarning mos elementlarini qo'shib bitta umumiy massiv qaytaradi.
```

```python
np.multiply(arr5,arr6) # multiply funksiyasi massivlarning mos elementlarini bir biriga ko'paytirib qaytaradi.
```

```python
np.maximum(arr5,arr6) # maximum funksiyasi massivlarning mos elementlarini solishtirib ularning kattasini qaytaradi.
```

subtrack, power, mod, copysign funksiyalari haqida gaplashamiz.

```python
arr7 = np.array([3,4,5])
arr8 = np.array([7,9,12])
```

```python
np.subtract(arr8,arr7) # birinchi massiv elementlaridan ikkinchi massiv elementlarini ayiradi.
```

```python
np.power(arr8,arr7)
```

```python
np.power([3,5,4],3) # bu funksiya birinchi elementning ikkinchi elementinchi darajasini hisoblaydi.
```

```python
np.mod(arr8,arr7)
```

```python
np.mod(17,6) # bu funksiya birinchi elementni ikkinchi elementga bo'ladi va qoldiqni qaytaradi.
```
