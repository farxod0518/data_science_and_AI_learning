### Mantiqiy shart operatorlari.

```python
import numpy as np
```

where - funksiyasi.

```python
arr1 = np.arange(1,11)
arr2 = np.arange(21,31) # bu yerda 2 ta turlicha massiv yaratdik.
```

```python
condition = np.array([True,False,False,True,False,False,True,False,True,True])
condition # bu yerda boolen malumot turidan iborat massiv yaratdik kelajakda shart aperatorida ishlatish uchun.
```

```python
natija = np.where(condition,arr1,arr2) # bu yerda funksiya conditionning har bir elementini o'qiyda agar True bo'lsa arr1 dan mos element oladi agar
                                       #  False bo'lsa arr2 dan oladi va yangi massiv qaytaradi. bu yerda 3 ta massivning ham element soni teng bo'lishi kerak.
```

```python
natija
```

```python
arr3 = np.random.randn(5,5)
arr3
```

```python
np.where(arr3<0, arr3*(-1), arr3) # bu yerda arr3 massivimizning barcha elementlarini musbat qilib olyapmiz.
```

```python
np.fabs(arr3) # barcha elementlarnining modulini qaytaruvchi funksiya
```

```python
np.where(np.fabs(arr3)<0.1, 0,arr3) # bu yerda elementlarning mpduli 0.1 dan kichik bo'lgan elementlarni 0 ga tenglab yubordik.
```

```python
np.where((arr3<0.1)&(arr3>(-0.1)), 0, arr3) # bu yerda xuddi shu ishni fabs funksiyasidan foydalanmasdan amalga oshirdik.
```

```python
arr3
```

```python
qator, ustun =np.where(arr3>0)
```

```python
print(qator)
print(ustun)
```

```python
joylashuv = np.array([qator,ustun])
joylashuv
```

```python
aniqjoylashuv = joylashuv.T # bu yerda arr3 massivimizdagi 0 dan katta bo'lgan elementlarimizning indexlarini ajratib aniqjoylashuv nomli massivimizga saqladik.
print(aniqjoylashuv)
```

```python
musbat_sonlar = np.array([])
for i in aniqjoylashuv: # bu yerda aniqjoylashuv massivimizdagi elementlardan foydalanib arr3dagi musbat soblarni ekranga chiqardik.
  a = i[0]
  b = i[1]
  musbat_sonlar = np.append(musbat_sonlar, arr3[a,b])  # qo'shimchasiga barcha musbat sonlarni alohida massivga yuklab oldik.
  print(arr3[a,b])
```

```python
print(musbat_sonlar)
```

## matematik va statistik usullardan foydalanish.

```python
arr7 = np.array([1,2,4,6,9])
```

## Tartiblash (sorting)
