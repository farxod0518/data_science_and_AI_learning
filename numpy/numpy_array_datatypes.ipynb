## massivlarda indexlar bilan ishlash.

```python
arr4 = np.arange(5,30,2,dtype=np.int16)
arr4
```

```python
arr4[3] # index orqali 1 ta malumot olganimizda malumot turini ham ko'rishimiz mumkin.
```

```python
arr4[3:7] # elementlarni indexlari orqali kesib olish.
```

```python
sliced = arr4[5:8]
sliced
```

```python
sliced[1]=1211 # bu yerda index orqali massivning malum elementini o'zgartirdik.
sliced
```

```python
arr4 # bu yerda ikkala o'zgaruvchi ham bir manbaga bog'langanligi sababli ikkala massivda ham o'zgarish yuz berdi.
```

copy() metodidan foydalanish.

```python
sliced_arr4 = arr4[0:4].copy()
sliced_arr4
```

```python
sliced_arr4[:] = 0
sliced_arr4
```

```python
arr4 # copy metodidan foydalanganligimiz sababli ikki massiv alohida malumotga bog'landi natijada biridagi o'zgarish ikkinchisiga tasir qilmadi.
```

boolen bilan indexlash.

```python
ismlar = np.array(['sardor','elyor','mirkomil','sardor','islom'])
```

```python
arr7 = np.random.randint(9,size=(5,4))
arr7
```

```python
ismlar == 'sardor' # bu yerda kiritgan malumot ismlar massivining qaysi elementi bilan bir
# xil bo'lsa shu element indexga True qolgan indexlarga esa False qiymat kiritib qaytaradi.
```

```python
arr7[ismlar=='sardor'] # bu yerda boolen dan foydalanib arr7 dagi malumotlarni chaqiryapmiz.
```

```python
arr7[ismlar!='sardor']
```

```python
arr7[~(ismlar=="sardor")]
```

```python
arr7[ismlar=='sardor']
```

```python
arr7[ismlar=='sardor'][:,1:3]
```

```python
arr7[ismlar=='sardor',2:3]
```

```python
arr7[(ismlar=="elyor") | (ismlar=='islom')]  # (|) or aperatori. bir vaqtda ikki elementga tegishli malumotlarni chiqaryapmiz.
```

```python
arr7
```

```python
arr7[arr7<4]=0 # 4 dan kichik barcha elementlarni 0 ga tengladik.
arr7
```

reshape funksiyasi.

```python
arr7 = np.full((1,9),1)
arr7
```

```python
re_arr7 = arr7.reshape((3,3))# bu yerda arr7 massivining shaklini o'zgartirdik.
re_arr7
```

```python
arr8 = np.reshape(np.arange(9),(3,3)) # bu yerda reshape funksiyasi yordamida o'zimiz hoxlagan shakldagi massive yaratdik.
arr8
```

```python
arr9 = np.reshape(arr8,(1,9))
arr9
```

Transpose va swapping axes metodlari.

```python
arr10 = np.arange(9).reshape(3,3)
arr10
```

```python
arr10.T # bu yerda qatorlarni ustunlarga ustunlarni esa qatorlarga almashtirdik.
```

```python
arr10.swapaxes(1,0)
```

```python
arr10.swapaxes(0,1)
```

dot metodi. massivlarni bir biriga ko'paytirishda ishlatiladi.

```python
np.dot(arr10,arr10.T)
```
