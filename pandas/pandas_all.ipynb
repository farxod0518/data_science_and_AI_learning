### Arifmetik amallar.

```python
import pandas as pd
import numpy as np
```

```python
ser1 = pd.Series([7.3, -3.5, 2.4, 1.4], index=['a', 'c', 'd', 'e'])
ser2 = pd.Series([-2.6, 4.6, -1.5, 4.2, 3.3], index=['a', 'c', 'e', 'f', 'g'])
```

```python
ser1
```

```python
ser2
```

```python
ser1+ser2 # 2 ta seriesni bir biriga qo'shganimizda mos elementlar bir biriga qo'shildi.
```

```python
df1 = pd.DataFrame(np.arange(9.).reshape(3,3), columns=list('bcd'), index=['Olma','Anor','Uzum'])
df2 = pd.DataFrame(np.arange(12.).reshape(4,3), columns=list('abc'), index=['Olma','Anor','Qovun','Anjir'])
```

```python
df1
```

```python
df2
```

```python
df1+df2 # bu yerda ham xuddi serieslardagidek qo'shish amali bajarildi. mos elementlar bir biriga qo'shildi va qolgan joylar nan qiymatini oldi.
```

```python
df1.add(df2, fill_value=0) # add methodidan foydalanib ikki dataframeni bir biniga qo'shdik va mavjud bo'lmagan elementlarini o bilan to'ldirdik.
                           # faqat ikkalasida ham mavjud bo'lmagan joylarni nan bilan to'ldirdik.
```

```python
df2- df1
```

```python
df2.sub(df1, fill_value=0) # ikki dataframelarni bir biridan ayirdik.
```

```python
df2.rsub(df1, fill_value=0)
```

```python
df2.div(df1) # bo'lish methodini ishlatdik.
```

```python
df2.floordiv(df1, fill_value=1) # bo'ldik va butun qismini oldik.
```

```python
df2.mul(df1, fill_value=1) # ko'paytirish methodini ishlatdik.
```

```python
df2.pow(2) # darajaga oshirdik.
```

### Obyektning elementlariga funksiyalarni qo'llash.

```python
df3 = pd.DataFrame(np.random.randn(3,4), index=['a','b','c'], columns=['Olma','Anor','Uzum','Qovun'])
df3
```

```python
np.abs(df3) # barcha elementlarning ishoralarini + ga o'zgartirdik.
```

```python
np.round(df3) # barcha elementlarni yaxlitladik.
```

```python
df3.apply(np.abs)
```

```python
df3.apply(sum, axis=1)
```

```python
l = lambda x:x.max()-x.min()
```

```python
df3.apply(l)
```

```python
def make_ser(a):
  return pd.Series([a.max(), a.min()], index=['max', 'min'])
```

```python
df3.apply(make_ser)
```

```python
df3.applymap(lambda x:x**3) # bu yerda har bir element uchun funksiyani qo'lladik.
```

## reytinglash(ranking) va tartiblash(sorting)

```python
ser5 = pd.Series([2,4,1,9,7], index=["c",'b','a','e','d'])
ser5
```

```python
ser5.sort_index() # series abyektimizni indexlari bo'yicha tartibladik.
```

```python
ser5.sort_values() # series abyektimizni qiymatlari bo'yicha tartibladik
```

```python
df5 = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/usa_cars.csv",index_col=0) # internetdan dataframe yuklab oldik.
df5.head()
```

```python
df5.shape
```

```python
df5.sort_index()
```

```python
df5.sort_index(axis=1) # dataframeni qatorlar bo'yicha tartibladik.
```

```python
df5.sort_index(ascending=False) # teskari tartibda tartibladik.
```

```python
df5.sort_values(by='price').head(20)
```

```python
df5.sort_values(by=['price','year'], ascending=False).head(20)
```

```python
df5['rating']=df5['mileage'].rank(method='average') # mashinalarni yurgan kilometri bo'yicha reytingladik.
df5.sort_values('rating')
```

```python
df5.info() # dataframe haqida qisqacha malumot.
```

```python
df5.describe()
```

### Korrellatsiya

```python
df5['price'].corr(df5['year']) # price va year ustunlarining bir-biri bilan bog'liqligini aniqladik.
```

```python
df5.corr(numeric_only=True) # barcha number qatorlarni solishtirdik.
```

```python
df5.corrwith(df5['year'],numeric_only=True) # bitta ustunni qolganlari bilan solishtirdik.
```
